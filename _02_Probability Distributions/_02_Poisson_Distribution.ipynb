```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import poisson
```

```python
np.random.poisson(lam=3.6)
```

```python
np.random.poisson(lam=3.6,size=10)
```

```python
sns.set_theme()
queue_size = np.random.poisson(lam=3.6,size=1000)
sns.countplot(x=queue_size)
plt.show()
```

### Using **SciPy**

```python
poisson.pmf(7,3.6)
```

```python
# poisson.cdf(k,mu) - Cumulative distribution function - for less than or equal to 2
# poisson.pmf(k,mu) - Probability mass function - for exact two defects
# poisson.sf(k,mu) - for more than 2 ( similar to 1 -cdf)
# poisson.mean(mu) - for mean of the distribution
# poisson.var(mu) - for variance of the distribution
# poisson.std(mu) - for standard deviation of the distribution

# poisson.mean(mu) = poisson.var(mu)
```

```python
# probability of less than 7 people in the queue.
poisson.cdf(7,3.6)
```

```python
# probability of more than 7 people in the queue.
poisson.sf(7,3.6)
```

```python
poisson.cdf(7,3.6) + poisson.sf(7,3.6)
```

```python
poisson.mean(3.6)
```

```python
poisson.var(3.6)
```

```python
poisson.std(3.6)
```

```python
x_ax = np.arange(0,11)
x_ax
```

```python
y_ax = poisson.pmf(x_ax,3.6)
```

```python
sns.barplot(x=x_ax,y=y_ax)
plt.xlabel('number of people in the queue')
plt.ylabel('Probability')
plt.title('Poisson Probability')
plt.show()
```

```python
y_ax = poisson.cdf(x_ax,3.6)
```

```python
# Cumulative plot
sns.barplot(x=x_ax,y=y_ax)
plt.xlabel('number of people in the queue')
plt.ylabel('probability')
plt.title('Cumulative probability plot')
plt.show()
```
