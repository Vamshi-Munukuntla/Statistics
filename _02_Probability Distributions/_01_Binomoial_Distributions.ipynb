# SciPy -> Stats

```python
from scipy.stats import binom
import numpy as np
```

```python
# Probability of getting two or less defectives
binom.cdf(k=2,n=20,p=0.12)
```

```python
# Probability of getting exactly 2 defectives
binom.pmf(k=2,n=20,p=0.12)
```

```python
# Mean of the binomial distribution
binom.mean(n=20,p=0.12)
```

```python
# Variance of the binomial distribution
binom.var(n=20,p=0.12)
```

```python
# Standard Deviation of the binomial distribution
binom.std(n=20,p=0.12)
```

```python
# Probability of getting more than 2 defectives
# Survival Function
binom.sf(k=2,n=20,p=0.12)
```

```python
# binom.cdf(k,n,p) - Cumulative distribution function - for less than or equal to 2
# binom.pmf(k,n,p) - Probability mass function - for exact two defects
# binom.sf(k,n,p) - for more than 2 ( similar to 1 -cdf)
# binom.mean(n,p) - for mean of the distribution
# binom.var(n,p) - for variance of the distribution
# binom.std(n,p) - for standard deviation of the distribution
```

```python
x_ax = np.arange(0,21)
```

```python
x_ax
```

```python
y_ax = binom.pmf(k=x_ax,n=20,p=0.12)
```

```python
c_ax = binom.cdf(k=x_ax,n=20,p=0.12)
```

```python
sns.barplot(x=x_ax,y=c_ax)
plt.show()
```
