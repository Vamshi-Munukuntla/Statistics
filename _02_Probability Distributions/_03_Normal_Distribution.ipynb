```python
# norm.cdf(x,mu, sigma) - Cumulative distribution function - for less than 
# norm.pdf(x,mu,sigma) - Probability density function (not Probability mass function)
# nomr.sf(x,mu,sigma) - for more than (similar to 1-cdf)
# norm.mean(mu)
# norm.var(mu)
# norm.std(mu)
```

```python
x_ax = np.linspace(144,156,13)
```

```python
y_ax = norm.pdf(x_ax,150,2)
```

```python
sns.lineplot(x=x_ax,y=y_ax)
plt.show()
```

```python
# Smoothening the curve by increasing x points
x_ax = np.linspace(144,156,13000)
y_ax = norm.pdf(x_ax,150,2)
sns.lineplot(x=x_ax,y=y_ax)
plt.axvline(153,color='grey')
plt.show()
```
