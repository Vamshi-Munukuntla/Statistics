# One sample Z test using Python

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import scipy.stats as stats
from statsmodels.stats import weightstats
```

```python
df = pd.read_csv('1.1 Machine1.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
sns.displot(data=df,x='Machine 1')
plt.show()
```

```python
sns.catplot(data=df, y='Machine 1',kind='box')
plt.show()
```

```python
df['Machine 1']
```

#### Performing Z-test

```python
weightstats.ztest(x1=df['Machine 1'], value=150, alternative='two-sided')
```

```python
 weightstats.ztest(x1=df['Machine 1'], value=150, alternative='larger')
```

#### Performing t-test

```python
stats.ttest_1samp(df['Machine 1'],popmean=150)
```
