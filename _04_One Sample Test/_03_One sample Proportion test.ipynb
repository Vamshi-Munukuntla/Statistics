```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
# Two tail test

stats.binom_test(x=14,n=100,p=0.21,alternative='two-sided')
```

```python
stats.binomtest(k=14,n=100,p=0.21,alternative='two-sided')
```

```python
# One tail test

stats.binom_test(x=14,n=100,p=0.21,alternative='less')
```
