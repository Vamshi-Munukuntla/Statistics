## One Variance Test using Python

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import scipy.stats as stats
```

```python
x_ax = np.linspace(0,100,101)
y_ax = stats.chi2.pdf(x_ax,df=50)

sns.relplot(x=x_ax,y=y_ax,kind='line')
plt.show()
```

Chi_sq_cal = (n-1)s^2 / sigma^2

```python
Chi_sq_cal = (51-1)*(2.35**2) / (2**2)
```

```python
Chi_sq_cal
```

```python
Chi_sq_critical = stats.chi2.isf(0.10,df=50)
```

```python
Chi_sq_critical > Chi_sq_cal
```

#### Hence we reject the Null Hypothesis.
