## Two Sample Z - Test using Python

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from statsmodels.stats import weightstats
import scipy.stats as stats
```

```python
df = pd.read_csv('1.1 Two Machines.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.set_theme()
sns.boxplot(data=df,x='Machine',y='Volume')
plt.show()
```

```python
sns.catplot(data=df,x='Machine',y='Volume',kind='box')
plt.show()
```

```python
m1 = df[df['Machine'] == 'Machine 1']['Volume']
```

```python
m2 = df[df['Machine'] == 'Machine 2']['Volume']
```

```python
### Two sample z-test

weightstats.ztest(m1,m2,alternative='two-sided')
```

```python
# Independent T-test (Two sample T-test)

stats.ttest_ind(m1,m2,equal_var=True)
```
