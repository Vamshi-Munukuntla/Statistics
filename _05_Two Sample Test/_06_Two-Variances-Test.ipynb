### Bartlett Test

```python
m1 = [150,150,151,148,151,152,149,150]
m2 = [135,152,154,169,157]
```

```python
np.std(m1)
```

```python
np.std(m2)
```

```python
stats.bartlett(m1,m2)
```

- Bartlett test takes an assumption that the m1, and m2 are normally distributed.

### Levene Test

```python
stats.levene(m1,m2)
```

- Levene test is truly a robust test.
- It doesn't take any assumptions.
- If the pvalue is less than 0.05, then we can believe that we can reject the null hypothesis.
