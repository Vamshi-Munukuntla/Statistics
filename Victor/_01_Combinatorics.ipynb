```python
def normal_dist(mu=14, sigma=2.8, N=5000):
    mu, sigma = mu, sigma
    N = N
    dist = np.random.normal(mu, sigma, N)
    sns.kdeplot(dist, color='black', linewidth=1.5)
    plt.axvline(mu, color='red')
    plt.axvline(mu+sigma, color = 'green')
    plt.axvline(mu-sigma, color='green')
    plt.show()
```

## PMF's 

```python
import matplotlib.pyplot as plt
import collections
import numpy as np
```

```python
roll_die1 = np.random.randint(1, 7, size=(3000))
roll_die2 = np.random.randint(1, 7, size=(3000))
sum_of_dice = roll_die1 + roll_die2

print(roll_die1)
print(roll_die2)
print(sum_of_dice)
```

```python
counter_die1 = collections.Counter(roll_die1)
counter_die2 = collections.Counter(roll_die2)
counter_sum  = collections.Counter(sum_of_dice)
print(f'The list of events: {sum_of_dice}')
print(counter_sum)
```

```python
plot_range = range(1,13)
```

```python
for k in plot_range:
    num = counter_die1.get(k,0)
    den = len(roll_die1)
    print(round(num/den,0))
```

```python
pmf_die1 = [round(counter_die1.get(k,0)/len(roll_die1), 2) for k in plot_range]
pmf_die2 = [round(counter_die2.get(k,0)/len(roll_die2), 2) for k in plot_range]
pmf_sum = [round(counter_sum.get(k,0)/len(sum_of_dice),2) for k in plot_range]
```

```python
# Plotting parts

plt.style.use('fivethirtyeight')

plt.title('Probability Mass Function (Die #1))')
plt.stem(plot_range, pmf_die1)
plt.show()

plt.title('Probability Mass Function (Die #2))')
plt.stem(plot_range, pmf_die2)
plt.show()
```

```python
plt.title('Probability Mass Function for Dice Summed')
plt.stem(plot_range, pmf_sum)
plt.show()
```

```python
plt.hist(sum_of_dice, range(1,12))
plt.show()
```

```python
path = r"C:/Users/vamsh/Videos/Semis/ProbabilityAndStats/StatisticalDistributions/data/titanic.csv"
df = pd.read_csv(path)
```

```python
all_ages = df.Age.dropna()
age_mean = all_ages.mean()
print(f'There are {all_ages.size} people, average age is {age_mean :.2f}')
```

```python
number_of_ppl_in_sample = 5
sample = all_ages.sample(n=number_of_ppl_in_sample, random_state=27)
mean_s = sample.mean()
```

```python
mean_s
```

```python
calc_percent_error = lambda pop_mean, sample_mean: np.abs(sample_mean - pop_mean)/pop_mean
```

```python
percent_err = calc_percent_error(age_mean, mean_s)
```

```python
print(f'The sample mean was {mean_s:1f} with a percent error of {percent_err*100:.2f}  ')
```

### Repeat the samples

```python
number_of_ppl_in_sample=20
sample_means = []
for i in range(500):
    sample = all_ages.sample(n=number_of_ppl_in_sample)
    sample_means.append(sample.mean())

print(sum(sample_means)/len(sample_means))
plt.hist(sample_means, bins=100)
plt.show()
```
