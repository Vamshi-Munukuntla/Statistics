## Confidence Intervals

```python
import numpy as np
import pandas as pd
```

```python
titanic= pd.read_csv('data/titanic.csv')
```

```python
ages = titanic.Age.dropna()
```

```python
ages.shape
```

```python
sample = ages.sample(10, replace=True)
sample
```

```python
rows=100
columns=10
size = (rows, columns)
samples = np.random.choice(ages,size)
means = np.mean(samples, axis=1)
```

```python
def get_all_sample_means(data, n_samples=100,n=10):
    
    #
    samples = np.random.choice(data, size=(n_samples, n))
    means = np.mean(samples, axis=1)
    #
    return means
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
samples = get_all_sample_means(n_samples=10**3, data=ages)
```

```python
sns.displot(samples, kde=False, rug=True)
plt.axvline(ages.mean(), color='red')
```

```python
sns.distplot(ages, color='orange', kde=True, rug=True)
plt.axvline(ages.mean(), color='red')
```

```python
sample = titanic.Fare.sample(10, replace=True)
# sample = ages.sample(10, replace=True)
bs = get_all_sample_means(sample, n=len(sample), n_samples=10**4)
np.mean(bs)
```
