```python
a = np.random.normal(100,5,1000)
b = np.random.normal(80,5,1000)
bm_data = np.append(a,b)
np.random.shuffle(bm_data)
```

```python
sns.displot(bm_data)
plt.show()
```

```python
# 500 samples with sample size = 4
bm_4 = bm_data.reshape(500,4).mean(axis=1)
sns.displot(bm_4)
plt.show()
```

```python
# 50 samples with sample size = 40
bm_40 = bm_data.reshape(50,40).mean(axis=1)
sns.displot(bm_40)
plt.show()
```

```python
x = np.random.normal(100,5,100000)
y = np.random.normal(80,5,100000)
bm_data = np.append(x,y)
np.random.shuffle(bm_data)
```

```python
sns.displot(bm_data)
plt.show()
```

```python
# 50000 samples with sample size = 4
bm_4 = bm_data.reshape(50000,4).mean(axis=1)
sns.displot(bm_4)
plt.show()
```

```python
x = np.random.normal(100,5,1000000)
y = np.random.normal(80,5,1000000)
bm_data = np.append(x,y)
np.random.shuffle(bm_data)

# 50000 samples with sample size = 4
bm_40 = bm_data.reshape(50000,40).mean(axis=1)
sns.displot(bm_40)
plt.show()
```
